```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#loading directly from seaborn repository
df = sns.load_dataset("iris")
```

```python
#inspect the dataset
print(" Dataset Shape ")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}\n")

print(" Column Names ")
print(list(df.columns), "\n")

print(" First 5 Rows (.head()) ")
print(df.head(), "\n")

print(" Info Summary (.info())")
df.info()
print("\n")

print(" Descriptive Statistics (.describe()) ")
print(df.describe(), "\n")
```

```python
#aesthetic style for the plots
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(10, 8))
scatter_plot = sns.pairplot(df, hue="species", palette="Set2")
scatter_plot.fig.suptitle(
    "Scatter Plot Matrix (Relationships between Features)", y=1.02
)
plt.show()
```

```python
#histogram
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Histograms: Feature Value Distributions", fontsize=16)

features = df.columns[:-1]  # Exclude 'species' column
colors = ["#4C72B0", "#55A868", "#C44E52", "#8172B3"]

for i, col in enumerate(features):
    ax = axes[i // 2, i % 2]
    sns.histplot(df[col], kde=True, ax=ax, color=colors[i], bins=15)
    ax.set_title(f"Distribution of {col.replace('_', ' ').title()}")
    ax.set_xlabel("")

plt.tight_layout()
plt.show()
```

```python
#boxplot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Box Plots: Identifying Outliers by Species", fontsize=16)

for i, col in enumerate(features):
    ax = axes[i // 2, i % 2]
    # hue="species" ensures compliance with newer seaborn formatting
    sns.boxplot(
        x="species",
        y=col,
        data=df,
        ax=ax,
        palette="Pastel1",
        hue="species",
        legend=False,
    )
    ax.set_title(f"{col.replace('_', ' ').title()} by Species")
    ax.set_xlabel("")
    ax.set_ylabel("")

plt.tight_layout()
plt.show()
```
